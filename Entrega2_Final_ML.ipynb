## Anlisis Profundo del Resultado de Correlacin

Esta simple lista de 10 nmeros cuenta una historia muy poderosa sobre los datos y el negocio del banco.

---

### 1. El Descubrimiento del "Atajo" (El Engao del Modelo Fcil)

La Evidencia: CUPO_MX tiene una correlacin de 0.81 con CUPO_L1. Este valor es extremadamente alto. Es mucho ms alto que cualquier otra variable predictora "legtima".

La Interpretacin: Esto es una alerta roja gigante de Target Leakage (Fuga de Datos). Significa que CUPO_L1 y CUPO_MX no son independientes. El banco casi con seguridad usa una frmula interna muy similar (basada en el perfil de riesgo del cliente) para calcular ambos cupos. En la prctica, son casi la misma informacin presentada de dos formas diferentes.

La Implicacin para el Modelo: Si hubiramos usado CUPO_MX para predecir CUPO_L1, el modelo habra aprendido una regla trivial: "Si quieres saber el cupo nacional, solo multiplica el cupo internacional por un factor X". El R habra sido altsimo (como nuestro 0.78 inicial), pero el modelo sera completamente intil para el negocio, porque para un cliente nuevo del cual queremos estimar un cupo, tampoco conoceramos su cupo internacional!

---

### 2. La Identificacin de los "Predictores Reales"

La Evidencia: Despus de la enorme cada desde CUPO_MX, las siguientes variables ms correlacionadas son Antiguedad (0.52) y Edad (0.44).

La Interpretacin: Estos son los verdaderos pilares sobre los que el banco construye su confianza en un cliente. La correlacin positiva y fuerte (aunque no tan extrema como la de CUPO_MX) nos dice que:

La lealtad se premia: A mayor antigedad, mayor tiende a ser el cupo.

La estabilidad importa: A mayor edad (proxy de estabilidad laboral y financiera), mayor tiende a ser el cupo.

---

### 3. La Validacin de Otras Variables de Negocio

La Evidencia: Variables como TC (Nmero de Tarjetas de Crdito), Renta, y PagoNac_T12 tambin aparecen en el top 10 con correlaciones positivas.

La Interpretacin: Esto confirma que el comportamiento financiero (cuntos productos tienes, cunto ganas, cmo pagas) tambin influye de manera relevante, aunque en menor medida que la lealtad y la estabilidad del cliente.

## Optimizacin de Hiperparmetros - (Fase 4: Modelado Avanzado)

Tu Random Forest us los parmetros por defecto (`n_estimators=100`, etc.). La optimizacin consiste en **encontrar una mejor combinacin de estos "ajustes"** para exprimir el mximo rendimiento del modelo.

---

### Qu es?

Usaremos **`RandomizedSearchCV`**, una tcnica que prueba combinaciones aleatorias de hiperparmetros para encontrar la mejor, de manera ms eficiente que probando todas las combinaciones posibles.

---

### Por qu es importante?

Es la forma estndar de mejorar el rendimiento de tu mejor modelo. Puede darte esos **puntos extra de R** que marcan la diferencia.

### Accin

Aade este bloque de cdigo.

```python
from sklearn.model_selection import RandomizedSearchCV

# ==============================================================================
#  FASE 4 (CONTINUACIN): OPTIMIZACIN DE HIPERPARMETROS
# ==============================================================================

print("--- Iniciando Optimizacin de Hiperparmetros para Random Forest ---")

# 1. Definir la grilla de parmetros a probar
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10],
    'max_features': ['sqrt', 'log2', 1.0] # Cambiado de 'auto' a 1.0 para versiones recientes
}

# 2. Crear el objeto RandomizedSearchCV
# n_iter=20 significa que probar 20 combinaciones aleatorias. cv=3 es validacin cruzada de 3-folds.
rf_random = RandomizedSearchCV(
    estimator=RandomForestRegressor(random_state=42, n_jobs=-1),
    param_distributions=param_dist,
    n_iter=20,
    cv=3,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# 3. Ajustar el modelo a los datos (usar X_train_reg y y_train_reg)
rf_random.fit(X_train_reg, y_train_reg)

# 4. Mostrar los mejores parmetros encontrados
print("\nMejores hiperparmetros encontrados:")
print(rf_random.best_params_)

# 5. Evaluar el modelo optimizado
best_rf_model = rf_random.best_estimator_
y_pred_best = best_rf_model.predict(X_test_reg)

mae_best = mean_absolute_error(y_test_reg, y_pred_best)
rmse_best = np.sqrt(mean_squared_error(y_test_reg, y_pred_best))
r2_best = r2_score(y_test_reg, y_pred_best)

print("\n--- Resultados del Modelo Random Forest Optimizado ---")
print(f"  Mean Absolute Error (MAE): ${mae_best:,.2f}")
print(f"  Root Mean Squared Error (RMSE): ${rmse_best:,.2f}")
print(f"  R-squared (R): {r2_best:.4f}")
```

## Validacin Cruzada en RandomizedSearchCV

---

El requisito de **utilizar validacin cruzada** se implementa directamente en este bloque mediante el parmetro `cv=3` dentro de **RandomizedSearchCV**.

---

### Proceso

- Para cada una de las **20 combinaciones de hiperparmetros** que se prueban (`n_iter=20`), el algoritmo divide el set de entrenamiento en **3 pliegues (folds)**.  
- Se realiza un ciclo de **3 entrenamientos y evaluaciones**.  
- La **puntuacin** de cada combinacin es el **promedio de su rendimiento** en estas 3 validaciones.

---

### Beneficio

Este mtodo garantiza que la **seleccin final de los mejores hiperparmetros** sea **robusta y fiable**, mitigando el riesgo de que un buen resultado sea producto del **azar** por una nica y favorable divisin de los datos.

- **`max_depth`: 30, `max_features`: 1.0, `min_samples_split`: 2**  Estos parmetros indican que el modelo funciona mejor cuando es **profundo y complejo**.  
  - Permite que el rbol crezca mucho (`max_depth=30`).  
  - Considera **todas las variables en cada divisin** (`max_features=1.0`).

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle

# 1. Binarizar las etiquetas (necesario para ROC multiclase)
y_test_bin = label_binarize(y_test_clf, classes=[0, 1, 2, 3])
n_classes = y_test_bin.shape[1]

# 2. Obtener las probabilidades de prediccin
y_score = clf_model.predict_proba(X_test_clf)

# 3. Calcular la curva ROC y el AUC para cada clase
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# 4. Graficar todas las curvas ROC
plt.figure(figsize=(10, 8))
colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='Curva ROC de la clase {0} (AUC = {1:0.2f})'
             ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curvas ROC para el Modelo Random Forest Classifier')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

## Interpretacin del Grfico

Cada lnea de color representa una de las **clases objetivo**: `0`, `1`, `2`, `3`.

---

### Curvas del Grfico

Las curvas que se encuentran ms *pegadas* a la esquina superior izquierda del grfico indican un **mejor rendimiento del modelo** para esa clase especfica.

---

### Valor AUC

El valor **AUC (Area Under the Curve)** mostrado en la leyenda cuantifica qu tan bien el modelo es capaz de distinguir cada clase de las dems.  
Un **AUC** ms cercano a `1` significa una **mejor capacidad de discriminacin**.

---

### Conclusin

Como se puede observar, las clases **0 ("Bajo")** y **3 ("Muy Alto")** presentan los valores de **AUC** ms altos, lo que confirma y refuerza lo visto anteriormente en el *Classification Report*: el modelo tiene un **rendimiento ms slido y preciso** al identificar las **categoras extremas**.


```python
from sklearn.model_selection import RandomizedSearchCV

# ==============================================================================
#  FASE 4 (CLASIFICACIN): OPTIMIZACIN DE HIPERPARMETROS
# ==============================================================================
print("\n--- Iniciando Optimizacin para Random Forest Classifier ---")

# 1. Definir la grilla de parmetros
# Es muy similar a la de regresin
param_dist_clf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}

# 2. Crear el objeto RandomizedSearchCV
# Esta vez, el scoring es 'f1_weighted' para optimizar nuestro F1-score
clf_random = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42, n_jobs=-1),
    param_distributions=param_dist_clf,
    n_iter=20,
    cv=3,
    verbose=2,
    random_state=42,
    n_jobs=-1,
    scoring='f1_weighted'
)

# 3. Ajustar el modelo
clf_random.fit(X_train_clf, y_train_clf)

# 4. Mostrar los mejores parmetros
print("\nMejores hiperparmetros encontrados:")
print(clf_random.best_params_)

# 5. Evaluar el modelo optimizado
best_clf_model = clf_random.best_estimator_
y_pred_best_clf = best_clf_model.predict(X_test_clf)

print("\n--- Resultados del Modelo Random Forest Classifier Optimizado ---")
print(classification_report(y_test_clf, y_pred_best_clf))
```
