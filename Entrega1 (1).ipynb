### Modelo de regresion lineal

```python
import numpy as np
import matplotlib.pyplot as plt

# Filtrar valores no nulos para evitar errores
x = df['Edad']
y = df['Antiguedad']

# Asegurarse de que las longitudes coincidan
x, y = x.align(y, join='inner')

# Calcular la regresin lineal
coeffs = np.polyfit(x, y, deg=1)  # Ajustar una lnea (grado 1)
m, b = coeffs  # Pendiente (m) y ordenada al origen (b)

# Crear valores de prediccin para la lnea de regresin
y_pred = m * x + b

# Graficar la relacin entre Edad y Antigedad con la lnea de regresin
plt.scatter(x, y, label='Datos', alpha=0.7, color='blue')
plt.plot(x, y_pred, color='red', label=f'Regresin lineal (y = {m:.2f}x + {b:.2f})')
plt.title('Relacin entre Edad y Antigedad')
plt.xlabel('Edad')
plt.ylabel('Antigedad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()


```

## Anlisis de la relacin entre Edad y Antigedad

### Correlacin
- Existe una **correlacin positiva, pero no determinante** entre **edad** y **antigedad**.  
- La edad influye en la permanencia, pero **no es suficiente por s sola para predecirla**.

###  Dispersin de los datos
- La **nube de puntos es muy dispersa**, lo que indica que la **regresin lineal explica solo parcialmente la relacin**.  
- Se observan dos casos particulares:
  - **Clientes jvenes con mucha antigedad**  probablemente clientes que iniciaron su relacin bancaria desde muy temprano.  
  - **Clientes mayores con poca antigedad**  posibles nuevas adquisiciones recientes.

###  Outliers
- Existen **clientes con antigedad extremadamente alta (200300 meses, ~1725 aos)**.  
- Estos valores atpicos pueden **distorsionar la tendencia lineal** y deben ser analizados aparte.

```python
# Variables
variables = ['CUPO_MX', 'Cuentas', 'Antiguedad', 'Edad', 'CUPO_L1', 'Dualidad', 'TC']

# 1. Identificacin de valores nulos
print("Valores nulos iniciales por columna:")
print(df[variables].isnull().sum())

# 2. Identificacin de outliers usando Z-score (solo en las columnas numricas)
numerical_cols = df[variables].select_dtypes(include=np.number).columns
z_scores = np.abs(zscore(df[numerical_cols].dropna()))
outliers = (z_scores > 3).any(axis=1)  # Marcar filas con algn outlier
print(f"Nmero de outliers detectados: {outliers.sum()}")

# 3. Dimensiones del DataFrame original
print(f"Dimensiones del DataFrame original: {df.shape}")

# 4. Imputacin de valores nulos y reemplazo de outliers
df_clean = df.copy()

for col in numerical_cols:
    median_value = df_clean[col].median()
    if not pd.isna(median_value):  # Verificar que la mediana no sea NaN
        # Reemplazar outliers con la mediana (asegurando compatibilidad de tipo)
        df_clean.loc[outliers, col] = int(round(median_value))

# Imputar valores nulos con la mediana de cada columna numrica
df_clean = df_clean.fillna(df_clean.median(numeric_only=True))

# Recalcular outliers despus de la limpieza
z_scores_clean = np.abs(zscore(df_clean[numerical_cols].dropna()))
outliers_clean = (z_scores_clean > 3).any(axis=1)
print(f"Nmero de outliers detectados despus de la limpieza: {outliers_clean.sum()}")

# 5. Verificar si se han imputado correctamente los valores nulos
print("Valores nulos despus de la imputacin:")
print(df_clean[variables].isnull().sum())

# 6. Dimensiones del DataFrame limpio
print(f"Dimensiones del DataFrame limpio: {df_clean.shape}")
```
